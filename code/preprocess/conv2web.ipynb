```python
def text_to_markdown_table(text):
    lines = text.split('\n')
    header = "| 患者信息 | 数值/描述 |\n|---|---|"
    rows = ""
    for line in lines:
        if line:  # 检查行是否为空
            key, value = line.split('：', 1)  # 使用中文冒号分割
            rows += f"\n| {key} | {value} |"
    return header + rows
```
