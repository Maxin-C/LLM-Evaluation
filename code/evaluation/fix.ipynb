```python
def log2msgs(log_content):
    msgs = [{
        "role": "system",
        "content": "现在你是一名乳腺癌专科医师，你的患者在手术后联系你获取帮助，该患者的信息如下：" + log_content['vp_prompt'] + "\n请结合患者信息给出尽可能专业有效并且容易理解的反馈。"
    }]
    for data in log_content['conversation']:
        msgs.append({
            "role": "user" if data['role'] == 'virtual patient' else "assistant",
            "content": data['content']
        })
    return msgs
```
